## STEMMING

#### Stemming is the process of reducing a word to its word stem that affixes to suffixes and preffixes or to the roots of the words known as lemma

```python
### Classification probelm 
### comments the product is a positive review or negative review
### review ---> eating, eaten, eats 
```

```python
words = ['eating','eats','eaten','writing','written','writes','history','final','finalize']
```

```python
### PorterStemmer

from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"---> "+stemming.stem(word))
```

```python
stemming.stem('congratulation')
```

```python
stemming.stem('fairly')
```

```python
### RegexStemming

from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('writes')
```

```python
reg_stemmer.stem('capable')
```

```python
### SnowballStemming 

from nltk.stem import SnowballStemmer
```

```python
snowball = SnowballStemmer('english')
```

```python
snowball.stem('eating')
```

```python
for word in words:
    print(word+"----> "+snowball.stem(word))
```
